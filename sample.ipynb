```python
def scrape_project(url):
    sleep(0.5)  # be nice to the server
    r = requests.get(url, headers=headers)
    s = BeautifulSoup(r.text, "html.parser")

    def get_text_by_header(header):
        tag = s.find("h2", string=header)
        if tag and tag.find_next_sibling():
            return tag.find_next_sibling().get_text(strip=True)
        return ""
    title = s.select_one("#content h1")
    description = get_text_by_header("Description")
    research_area = get_text_by_header("Domaines de recherche")
    profile = get_text_by_header("Profil recherché")
    funding = get_text_by_header("Financement en lien avec le projet de recherche")
    contact_section = s.find("h2", string="Pour plus d'information")
    email = ""
    faculty = ""
    if contact_section:
        for sib in contact_section.find_all_next():
            text = sib.get_text(strip=True)
            if "@" in text and not email:
                email = text
            if "Faculté" in text and not faculty:
                faculty = text
            if email and faculty:
                break

    return {
        "Title": title.text.strip() if title else "",
        "Description": description,
        "General Research Area": research_area,
        "Research Interests / Skills": profile,
        "Funding Info": funding,
        "Email": email,
        "Faculty": faculty,
        "URL": url
    }
```
